```python
# ============================================================
# ANÁLISIS DE ACCIDENTES DE TRÁFICO
# Realizado por: Sergio Gómez García
#
# Métodos de IA utilizados: clasificación, regresión,
#   redes neuronales y clustering.
# ============================================================

# 0. IMPORTAMOS LOS PAQUETES
import numpy as np
import pandas as pd

from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    f1_score,
    mean_absolute_error,
    mean_squared_error
)

from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV
from sklearn.linear_model import LogisticRegression, PoissonRegressor
from sklearn.neural_network import MLPClassifier
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.metrics import silhouette_score

import lightgbm as lgb  # Modelo de gradiente boosting para clasificación multiclase


# 1. CARGA Y LIMPIEZA DE LOS DATOS

# Ruta CSV (Cambiar dependiendo del entorno utilizado)
DATA_PATH = "/content/drive/MyDrive/ACCIDENTES-DEFINITIVOS3.csv"

df = pd.read_csv(DATA_PATH, sep=";", low_memory=False)

# Normalización de cadenas: mayúsculas y sin espacios
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype(str).str.strip().str.upper()

print("Dimensión del conjunto de datos:", df.shape)
print("Variables disponibles:", df.columns.tolist())

# Comprobación de la distribución de la variable objetivo
print("\nDistribución de TIPO_ACCIDENTE (variable multiclase):")
print(df["TIPO_ACCIDENTE"].value_counts())

# Variables numéricas clave
vars_num = ["TOTAL_VICTIMAS_24H", "TOTAL_MU24H", "TOTAL_HG24H", "TOTAL_HL24H"]
vars_num = [v for v in vars_num if v in df.columns]

print("\nResumen de variables de severidad (24h):")
print(df[vars_num].describe())


# 2. FUNCIÓN AUXILIAR: MATRIZ DE CONFUSIÓN

def plot_confusion_matrix(y_true, y_pred, labels, title):
    """
    Dibuja una matriz de confusión normalizada por filas.
    Sirve para interpretar qué tipos de accidente se confunden más.
    """
    cm = confusion_matrix(y_true, y_pred, normalize="true")
    plt.figure(figsize=(10, 7))
    sns.heatmap(
        cm,
        annot=True,
        fmt=".2f",
        cmap="Blues",
        xticklabels=labels,
        yticklabels=labels
    )
    plt.title(title)
    plt.xlabel("Predicción")
    plt.ylabel("Real")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()


# 3. CLASIFICACIÓN MULTICLASE: TIPO_ACCIDENTE

# --- 3.1 Preparación de datos de entrada ---

# Eliminamos filas con TIPO_ACCIDENTE desconocido (si existen)
df_clf = df.dropna(subset=["TIPO_ACCIDENTE"]).copy()

X = df_clf.drop(columns=["TIPO_ACCIDENTE"])
y = df_clf["TIPO_ACCIDENTE"]

# Identificamos columnas categóricas y numéricas
cat_cols = X.select_dtypes(include=["object"]).columns.tolist()
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()

# Preprocesado: imputamos y codificamos variables categóricas + escalar variables numéricas
preprocess_clf = ColumnTransformer(
    transformers=[
        (
            "cat",
            Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("onehot", OneHotEncoder(handle_unknown="ignore"))
            ]),
            cat_cols
        ),
        (
            "num",
            Pipeline([
                ("imputer", SimpleImputer(strategy="median")),
                ("scaler", StandardScaler())
            ]),
            num_cols
        )
    ]
)

# Dividimos datos en entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# 3.2 MODELO BASE: Random Forest

rf_pipeline = Pipeline(
    steps=[
        ("preprocess", preprocess_clf),
        ("clf", RandomForestClassifier(
            n_estimators=200,
            max_depth=15,
            min_samples_split=4,
            random_state=42,
            n_jobs=-1
        ))
    ]
)

print("\nEntrenando Random Forest (multiclase TIPO_ACCIDENTE)...")
rf_pipeline.fit(X_train, y_train)
y_pred_rf = rf_pipeline.predict(X_test)

print("\n=== RANDOM FOREST ===")
print(classification_report(y_test, y_pred_rf))
print("Accuracy RF:", accuracy_score(y_test, y_pred_rf))

plot_confusion_matrix(
    y_test,
    y_pred_rf,
    labels=sorted(y.unique()),
    title="Matriz de confusión — Random Forest"
)

# 3.3 SVM LINEAL (LinearSVC)

svm_base = LinearSVC(
    C=1.0,
    random_state=42,
    max_iter=5000
)

svm_calibrated = CalibratedClassifierCV(
    estimator=svm_base,
    method="sigmoid",
    cv=3
)

svm_pipeline = Pipeline(
    steps=[
        ("preprocess", preprocess_clf),
        ("clf", svm_calibrated)
    ]
)

print("\nEntrenando SVM lineal calibrada (multiclase TIPO_ACCIDENTE)...")
svm_pipeline.fit(X_train, y_train)
y_pred_svm = svm_pipeline.predict(X_test)

print("\n=== SVM LINEAL (calibrada) ===")
print(classification_report(y_test, y_pred_svm))
print("Accuracy SVM:", accuracy_score(y_test, y_pred_svm))

plot_confusion_matrix(
    y_test,
    y_pred_svm,
    labels=sorted(y.unique()),
    title="Matriz de confusión — SVM lineal (calibrada)"
)

# 3.4 MODELO PRINCIPAL: LightGBM

lgbm_pipeline = Pipeline(
    steps=[
        ("preprocess", preprocess_clf),
        ("clf", lgb.LGBMClassifier(
            objective="multiclass",
            num_class=y.nunique(),
            learning_rate=0.08,
            n_estimators=300,
            num_leaves=64,
            subsample=0.9,
            colsample_bytree=0.9,
            random_state=42,
            n_jobs=-1
        ))
    ]
)

print("\nEntrenando LightGBM (multiclase TIPO_ACCIDENTE)...")
lgbm_pipeline.fit(X_train, y_train)
y_pred_lgb = lgbm_pipeline.predict(X_test)

print("\n=== LIGHTGBM ===")
print(classification_report(y_test, y_pred_lgb))
print("Accuracy LightGBM:", accuracy_score(y_test, y_pred_lgb))

plot_confusion_matrix(
    y_test,
    y_pred_lgb,
    labels=sorted(y.unique()),
    title="Matriz de confusión — LightGBM"
)

# Comparamos las métricas globales de ambos modelos
metrics_rf = {
    "Modelo": "Random Forest",
    "Accuracy": accuracy_score(y_test, y_pred_rf),
    "F1_weighted": f1_score(y_test, y_pred_rf, average="weighted")
}
metrics_svm = {
    "Modelo": "SVM lineal (calibrada)",
    "Accuracy": accuracy_score(y_test, y_pred_svm),
    "F1_weighted": f1_score(y_test, y_pred_svm, average="weighted")
}
metrics_lgb = {
    "Modelo": "LightGBM",
    "Accuracy": accuracy_score(y_test, y_pred_lgb),
    "F1_weighted": f1_score(y_test, y_pred_lgb, average="weighted")
}
df_metrics_clf = pd.DataFrame([metrics_rf, metrics_svm, metrics_lgb])
print("\nComparación global de modelos de clasificación:")
print(df_metrics_clf)


# 4. REGRESIÓN DE CONTEO: Nº DE VÍCTIMAS 24H

df_reg = df.copy()
df_reg = df_reg.dropna(subset=["TOTAL_VICTIMAS_24H"])

y_reg = df_reg["TOTAL_VICTIMAS_24H"]
X_reg = df_reg.drop(columns=["TOTAL_VICTIMAS_24H"])

cat_cols_reg = X_reg.select_dtypes(include=["object"]).columns.tolist()
num_cols_reg = X_reg.select_dtypes(include=[np.number]).columns.tolist()

preprocess_reg = ColumnTransformer(
    transformers=[
        (
            "cat",
            Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("onehot", OneHotEncoder(handle_unknown="ignore"))
            ]),
            cat_cols_reg
        ),
        (
            "num",
            Pipeline([
                ("imputer", SimpleImputer(strategy="median")),
                ("scaler", StandardScaler())
            ]),
            num_cols_reg
        )
    ]
)

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg,
    test_size=0.20,
    random_state=42
)

poisson_pipe = Pipeline(
    steps=[
        ("preprocess", preprocess_reg),
        ("reg", PoissonRegressor(alpha=1e-3, max_iter=300))
    ]
)

print("\nEntrenando PoissonRegressor para TOTAL_VICTIMAS_24H...")
poisson_pipe.fit(X_train_reg, y_train_reg)

y_pred_reg = poisson_pipe.predict(X_test_reg)

mae = mean_absolute_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))

# Métricas de regresión
metrics_reg = pd.DataFrame([{
    "Modelo": "PoissonRegressor",
    "MAE": mae,
    "RMSE": rmse
}])

print("\nTabla de métricas (regresión):")
print(metrics_reg)

# Observado vs Predicho
df_pred = pd.DataFrame({
    "y_real": pd.to_numeric(y_test_reg, errors="coerce"),
    "y_pred": pd.to_numeric(y_pred_reg, errors="coerce")
}).dropna()

# Forzamos tipos numéricos (clave)
df_pred["y_real"] = df_pred["y_real"].astype(int)
df_pred["y_pred"] = df_pred["y_pred"].astype(float)


tabla_obs_pred = (
    df_pred
    .groupby("y_real", as_index=False)
    .apply(lambda g: pd.Series({
        "n": len(g),
        "pred_media": g["y_pred"].mean(),
        "pred_mediana": g["y_pred"].median(),
        "pred_p25": np.percentile(g["y_pred"], 25),
        "pred_p75": np.percentile(g["y_pred"], 75),
        "mae_grupo": np.mean(np.abs(g["y_pred"] - g["y_real"]))  # <- aquí ya no hay strings
    }))
    .reset_index(drop=True)
    .sort_values("y_real")
)

print("\nTabla observado vs predicho (por víctimas reales):")
print(tabla_obs_pred.head(15))

# Coeficientes del modelo
ohe = poisson_pipe.named_steps["preprocess"].named_transformers_["cat"].named_steps["onehot"]
cat_feature_names = ohe.get_feature_names_out(cat_cols_reg)

num_feature_names = np.array(num_cols_reg)
feature_names = np.concatenate([cat_feature_names, num_feature_names])

coefs = poisson_pipe.named_steps["reg"].coef_

df_coefs = pd.DataFrame({
    "feature": feature_names,
    "coef": coefs,
    "abs_coef": np.abs(coefs)
}).sort_values("abs_coef", ascending=False)

print("\nTop 20 variables más influyentes:")
print(df_coefs.head(20)[["feature", "coef"]])

# Gráficos

print("\n=== REGRESIÓN DE CONTEO (Poisson) ===")
print("MAE (Error absoluto medio):", mae)
print("RMSE (Error cuadrático medio):", rmse)

plt.figure(figsize=(6, 6))
plt.scatter(y_test_reg, y_pred_reg, alpha=0.2)
plt.xlabel("Víctimas observadas (24h)")
plt.ylabel("Víctimas predichas (24h)")
plt.title("PoissonRegressor – Predicción nº de víctimas 24h")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


# 5. MODELO BINARIO DE GRAVEDAD (MUERTES > 0)

df_bin = df.copy()
df_bin["GRAVE"] = (df_bin["TOTAL_MU24H"] > 0).astype(int)

X_bin = df_bin.drop(columns=["GRAVE"])
y_bin = df_bin["GRAVE"]

cat_cols_bin = X_bin.select_dtypes(include=["object"]).columns.tolist()
num_cols_bin = X_bin.select_dtypes(include=[np.number]).columns.tolist()

preprocess_bin = ColumnTransformer(
    transformers=[
        (
            "cat",
            Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("onehot", OneHotEncoder(handle_unknown="ignore"))
            ]),
            cat_cols_bin
        ),
        (
            "num",
            Pipeline([
                ("imputer", SimpleImputer(strategy="median")),
                ("scaler", StandardScaler())
            ]),
            num_cols_bin
        )
    ]
)

X_train_bin, X_test_bin, y_train_bin, y_test_bin = train_test_split(
    X_bin, y_bin,
    test_size=0.20,
    random_state=42,
    stratify=y_bin
)

# 5.1 Regresión logística

logit_pipe = Pipeline(
    steps=[
        ("preprocess", preprocess_bin),
        ("clf", LogisticRegression(
            max_iter=500,
            n_jobs=-1
        ))
    ]
)

print("\nEntrenando Regresión Logística para GRAVE...")
logit_pipe.fit(X_train_bin, y_train_bin)
y_pred_log = logit_pipe.predict(X_test_bin)

print("\n=== LOGISTIC REGRESSION (GRAVE) ===")
print(classification_report(y_test_bin, y_pred_log))
print("Accuracy:", accuracy_score(y_test_bin, y_pred_log))

# 5.2 Red neuronal MLP

mlp_pipe = Pipeline(
    steps=[
        ("preprocess", preprocess_bin),
        ("clf", MLPClassifier(
            hidden_layer_sizes=(64, 32),
            activation="relu",
            solver="adam",
            random_state=42,
            max_iter=30,
            early_stopping=True
        ))
    ]
)

print("\nEntrenando MLPClassifier (red neuronal) para GRAVE...")
mlp_pipe.fit(X_train_bin, y_train_bin)
y_pred_mlp = mlp_pipe.predict(X_test_bin)

print("\n=== MLP (GRAVE) ===")
print(classification_report(y_test_bin, y_pred_mlp))
print("Accuracy:", accuracy_score(y_test_bin, y_pred_mlp))

# Comparamos ambos modelos binarios
metrics_bin = pd.DataFrame([
    {
        "Modelo": "Logistic Regression",
        "Accuracy": accuracy_score(y_test_bin, y_pred_log),
        "F1_weighted": f1_score(y_test_bin, y_pred_log, average="weighted")
    },
    {
        "Modelo": "MLP Neural Network",
        "Accuracy": accuracy_score(y_test_bin, y_pred_mlp),
        "F1_weighted": f1_score(y_test_bin, y_pred_mlp, average="weighted")
    }
])
print("\nComparación global de modelos binarios GRAVE:")
print(metrics_bin)


# 6. CLUSTERING DE PATRONES DE ACCIDENTE

df_clust = df.copy()

# Seleccionamos las variables para clustering
cols_cat_clust = [
    "MES",
    "DIA_SEMANA",
    "ZONA_AGRUPADA",
    "TITULARIDAD_VIA",
    "TIPO_VIA",
    "TRAZADO_PLANTA",
    "CONDICION_ILUMINACION",
    "CONDICION_METEO",
    "CONDICION_FIRME",
]

cols_cat_clust = [c for c in cols_cat_clust if c in df_clust.columns]

cols_num_clust = [
    "HORA",
    "TOTAL_VICTIMAS_24H",
    "TOTAL_MU24H",
    "TOTAL_HG24H",
    "TOTAL_HL24H"
]
cols_num_clust = [c for c in cols_num_clust if c in df_clust.columns]

X_clust = df_clust[cols_cat_clust + cols_num_clust].copy()

preprocess_clust = ColumnTransformer(
    transformers=[
        (
            "cat",
            Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
            ]),
            cols_cat_clust
        ),
        (
            "num",
            Pipeline([
                ("imputer", SimpleImputer(strategy="median")),
                ("scaler", StandardScaler())
            ]),
            cols_num_clust
        )
    ],
    remainder="drop"
)

# PCA a 10 componentes (reducimos dimensionalidad antes de aplicar KMeans)
pca = PCA(n_components=10, random_state=42)

pipe_pca = Pipeline(
    steps=[
        ("preprocess", preprocess_clust),
        ("pca", pca)
    ]
)

X_pca = pipe_pca.fit_transform(X_clust)
print("\nShape tras PCA (clustering):", X_pca.shape)

# Realizamos una búsqueda rápida del nº de clusters con silhouette
range_k = range(2, 8)
sil_scores = []

for k in range_k:
    kmeans_tmp = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_tmp = kmeans_tmp.fit_predict(X_pca)
    sil = silhouette_score(X_pca, labels_tmp)
    sil_scores.append(sil)
    print(f"k={k}, silhouette={sil:.3f}")

best_k = range_k[int(np.argmax(sil_scores))]
print("Número de clusters recomendado según silhouette:", best_k)

# Clustering final
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_pca)
df_clust["CLUSTER_ACCIDENTE"] = clusters

# Proyección 2D para visualización
pca_2d = PCA(n_components=2, random_state=42)
X_2d = pca_2d.fit_transform(pipe_pca.named_steps["preprocess"].transform(X_clust))

plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_2d[:, 0],
    X_2d[:, 1],
    c=clusters,
    cmap="Set1",
    s=10,
    alpha=0.4
)
plt.title("Clustering de accidentes (PCA 2D + KMeans)")
plt.xlabel("PC1")
plt.ylabel("PC2")
handles, _ = scatter.legend_elements()
plt.legend(handles, [f"Cluster {i}" for i in range(best_k)], title="Cluster")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Resumen de los perfiles del cluster
print("\nDistribución de accidentes por cluster:")
print(df_clust["CLUSTER_ACCIDENTE"].value_counts(normalize=True).rename("proporción"))

# Distribución de los tipos de accidente por cluster
if "TIPO_ACCIDENTE" in df_clust.columns:
    tabla_tipo_cluster = (
        df_clust
        .groupby("CLUSTER_ACCIDENTE")["TIPO_ACCIDENTE"]
        .value_counts(normalize=True)
        .rename("proporción")
        .groupby(level=0)
        .head(5)
    )
    print("\nPrincipales tipos de accidente por cluster (proporciones):")
    print(tabla_tipo_cluster)

# Nivel de gravedad media por cluster
sev_cols = [c for c in ["TOTAL_VICTIMAS_24H", "TOTAL_MU24H", "TOTAL_HG24H", "TOTAL_HL24H"]
            if c in df_clust.columns]
if sev_cols:
    print("\nNivel de gravedad media por cluster:")
    print(df_clust.groupby("CLUSTER_ACCIDENTE")[sev_cols].mean())
```
